```python
# Import libraries for visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
titanic = sns.load_dataset('titanic')

# Number of samples, number of attributes
titanic.shape
```

```python
# Information about the dataset 
# info() - concise summary of DataFrame.
titanic.info()
```

```python
# Look at first 5 items
titanic.head()
```

```python
# Describe data - measures of central tendency and dispersion

# Describe the quantitative data
titanic.describe()
```

```python
# Describe the categorical data
titanic.describe(include=object)
```

```python
# Create a histogram of the age of people
sns.histplot(data=titanic, x="age")
```

```python
sns.histplot(data=titanic, x="alive")
```

```python
# Count plot: Displays the count of observations in each category
# Data shows most people died during titanic
sns.countplot(x="alive", data=titanic)
plt.show()

# Good practice to write a sentence from what you observe in a graph. 
```

```python
# Plot bivariate data

sns.countplot(x="survived", hue="sex", data=titanic)
plt.show()

# More females survived
```

```python
# Class and survived bivariate analysis
sns.countplot(x="survived", hue="class", data=titanic)
plt.show()
```

```python
# scatter plot - relationship between continuous variables
sns.scatterplot(x="fare", y="age", data=titanic)
```

```python
# Heat map - visualize data in a matrix format, values are represented by color 
numeric_titanic = titanic.select_dtypes(include=["float64", "int64"])

sns.heatmap(numeric_titanic.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```

```python
# Combining categorical and numerical data
sns.boxplot(x="sex", y="age", data=titanic)
```

```python
# Box plot for class and age
sns.boxplot(x="class", y="age", data=titanic)
```

```python
# Swarmplot
sns.swarmplot(x="class", y="age", data=titanic)
```
