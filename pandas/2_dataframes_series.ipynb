```python
import pandas as pd
```

```python
# Read DataFrames from CSV files
df = pd.read_csv("data/survey_results_public.csv")
schema_df = pd.read_csv("data/survey_results_schema.csv")
```

```python
# Set options for pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
# The key is the columns of a data frame, and the values correspond to the values of each row
# When converted to a DataFrame, each value in the rows will be indexed.
people = {
    "first": ["Corey", "Jane", "John"],
    "last": ["Schafer", "Doe", "Doe"],
    "email": ["CoreyMSchafer@gmail.com", "JaneDoe@email.com", "JohnDoe@email.com"]
}

people
```

```python
# Generating a DataFrame from the dictorionary with keys as column names
# If all arrays are not the same size, an error will be thrown.
# This is because each array value will be indexed and must match.
people_df = pd.DataFrame(people)
people_df
```

```python
'''
Can see that each value is indexed using object RangeIndex.
'''

people_df.to_dict()
```

```python
'''
Retrieve a column of a DataFrame. Can retrieve either using
key notation (df.[col_name]) or dot notation (df.col_name).

Example:
- people_df['email']
- people_df.email

'''
people_df['email']
```

```python
'''
The type of each column is a series (1-D data).
- Simply put: It's all rows of a single column.
'''

type(people_df['email'])
```

```python
'''
Retrieving multiple columns. Requires using a nested list.
- When retrieiving multiple columns, we are filtering out other columns 
and returning a DataFrame.

Example: df[[col1, col2]] 
'''

people_df[['last', 'email']]
```

```python
'''
df.columns: Allows us to get a list of all the columns in our DataFrame.
'''

people_df.columns
```

```python
'''
Retrieving a Row:
- iloc (integer location): Retrieve row by the integer location.
  - Example: df.iloc[0] retrieves the first row of data (Series).
  - Example: df.iloc[[0, 1]] retrieves the first & second row of data (DataFrame).
  - Example: df.iloc[[0, 1], [0, 2]] retrieves the first two rows, and the first and third column.
    Can only specify the column indexes when using iloc. 
'''

people_df.iloc[[0, 1], [0, 2]]
```

```python
'''
Retrieving a Row:
- loc (location): Retrieves a row by the label (either the default RangeIndex or a custom index).
  - Example: people_df.loc[[0, 2]] retrieves the rows with index of 0 and 2.
    By default, RangeIndex is the index. But we can set custom index.
  - Example: people_df[[0, 2], 'email'] retrieves the row and column 'email'.
    For loc we can specify the label of the columns we want.
'''
people_df.loc[[0, 1], ['email', 'first']]
```

```python
'''
Counting values in a column.
- value_counts: Provides the value count for each value.

Example: df['RemoteWork'].value_counts()
'''

df['RemoteWork'].value_counts()
```

```python
'''
Retrieving a row: 
df.loc[0]: Retrieves the first persons response for all questions.
df.loc[0, 'RemoteWork']: Retrieves the first persons response for the RemoteWork column.

Using Slicing: Includes the first and last item. 
df.loc[0:100]: Will retrieve rows from 0 to 100.
df.loc[range(0, 100)]: Will do the same thing and retrieve rows form 0 to 100.
df.loc[0:100, 'Employment':'LearnCode']: Retrieves rows 0 to 100 (inclusive), 
                                        and columns 'Employment' to 'LearnCode' (inclusive).
'''

df.loc[0: 100, 'Employment':'LearnCode']
```
